# Introduction to Linear Algebra with SpectralSandbox

This notebook walks through the core operators and ISEP utilities provided by
`spectral_sandbox`.

```python
import numpy as np
import matplotlib.pyplot as plt
from spectral_sandbox import (
    diagonal_operator,
    random_hermitian,
    laplacian_operator,
    isep_check,
    isep_basis,
)
```

## 1. Diagonal Operators

```python
D = diagonal_operator([1, 2, 3, 4])
print(D)
passes, err = isep_check(D)
print(f"ISEP: {passes}, residual: {err:.2e}")
```

## 2. Random Hermitian (GUE) Operators

```python
H = random_hermitian(6, seed=0)
vals, vecs = isep_basis(H)
print("Eigenvalues:", np.round(vals, 4))
passes, err = isep_check(H)
print(f"ISEP: {passes}, residual: {err:.2e}")
```

## 3. Discrete Laplacian

```python
n = 20
L = laplacian_operator(n)
vals_L, _ = isep_basis(L)

plt.figure(figsize=(7, 3))
plt.plot(vals_L, 'o-')
plt.title(f'Eigenvalues of the {n}-site Laplacian')
plt.xlabel('Index k')
plt.ylabel('λ_k')
plt.tight_layout()
plt.show()
```

## 4. Level-Spacing Distribution (GUE)

For large GUE matrices the normalised nearest-neighbour spacings follow the
Wigner surmise `p(s) ≈ (π/2) s exp(−π s²/4)`.

```python
N = 300
H_big = random_hermitian(N, seed=1)
eigvals = np.linalg.eigvalsh(H_big)
spacings = np.diff(np.sort(eigvals))
spacings /= spacings.mean()  # unfold

s = np.linspace(0, 4, 200)
wigner = (np.pi / 2) * s * np.exp(-np.pi * s**2 / 4)

plt.figure(figsize=(6, 4))
plt.hist(spacings, bins=50, density=True, alpha=0.6, label='GUE sample')
plt.plot(s, wigner, 'r-', lw=2, label='Wigner surmise')
plt.xlabel('Normalised spacing s')
plt.ylabel('p(s)')
plt.title('GUE Level-Spacing Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```
